```python
import pandas as pd

# 'data.csv' नावाची फाइल लोड करा.
# तुमच्या फाइलचे नाव इथे लिहा.
df = pd.read_csv('mushrooms.csv')

# डेटा फ्रेम (dataframe) प्रिंट करा.
print(df)
```

```python
import numpy as np

# '?' ला NaN मध्ये बदलणे
df.replace('?', np.nan, inplace=True)

# 'stalk-root' कॉलमच्या Mode (सर्वात वारंवार येणाऱ्या मूल्याने) NaN भरणे
stalk_root_mode = df['stalk-root'].mode()[0]
df['stalk-root'].fillna(stalk_root_mode, inplace=True)

# 'veil-type' कॉलम वगळणे
df.drop('veil-type', axis=1, inplace=True)

print(df.info()) # नवीन डेटासेटची माहिती तपासा
```

```python
# 'class' (लक्ष्य/Target) कॉलम वेगळा करणे.
# हा 'e' (खाण्यायोग्य/edible) आणि 'p' (विषारी/poisonous) असा बायनरी आहे.
y = df['class']
X = df.drop('class', axis=1) # फीचर्स (Features)

# वन-हॉट एन्कोडिंग लागू करणे
X_encoded = pd.get_dummies(X)

# लक्ष्य (Target) कॉलमचे Label Encoding (e=0, p=1) करणे
y_encoded = y.map({'e': 0, 'p': 1})

# नवीन डेटासेटचा आकार तपासा (Features)
print("\nOne-Hot Encoding नंतर फीचर्सचा आकार:", X_encoded.shape)
print("\nपहिली ५ एन्कोडेड फीचर्स (Encoded Features):")
print(X_encoded.head())

# y_encoded ची पहिली ५ मूल्ये
print("\nपहिली ५ एन्कोडेड लक्ष्य (Encoded Target) मूल्ये:")
print(y_encoded.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import joblib  # 'joblib' लायब्ररी इंपोर्ट करणे

# १. डेटा क्लीनिंग आणि फीचर इंजिनियरिंग (Data Preprocessing)
# डेटा लोड करा
df = pd.read_csv("mushrooms.csv")

# '?' ला NaN मध्ये बदलून 'stalk-root' मध्ये Mode (सर्वात वारंवार येणाऱ्या मूल्याने) भरणे
df.replace('?', np.nan, inplace=True)
stalk_root_mode = df['stalk-root'].mode()[0]
df['stalk-root'].fillna(stalk_root_mode, inplace=True)

# 'veil-type' कॉलम वगळणे
df.drop('veil-type', axis=1, inplace=True)

# लक्ष्य (Target) आणि फीचर्स (Features) वेगळे करणे
y = df['class']
X = df.drop('class', axis=1)

# वन-हॉट एन्कोडिंग लागू करणे (One-Hot Encoding)
X_encoded = pd.get_dummies(X)

# लक्ष्य (Target) कॉलमचे Label Encoding (e=0, p=1) करणे
y_encoded = y.map({'e': 0, 'p': 1})

# २. डेटा स्प्लिट (Data Split)
# डेटा ट्रेनिंग (80%) आणि टेस्टिंग (20%) सेटमध्ये विभाजित करणे
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42
)

# ३. मॉडेल ट्रेनिंग (Model Training)
# Logistic Regression मॉडेल तयार करणे आणि ट्रेनिंग डेटावर ट्रेन करणे
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train, y_train)

# ४. मॉडेल इव्हॅल्युएशन (Model Evaluation)
# टेस्ट डेटावर प्रेडिक्शन (Prediction) करणे
y_pred = model.predict(X_test)

# अचूकता (Accuracy) आणि वर्गीकरण अहवाल प्रिंट करणे
print("मॉडेलची अचूकता (Accuracy):", accuracy_score(y_test, y_pred))
print("\nवर्गीकरण अहवाल (Classification Report):\n", classification_report(y_test, y_pred))

# ५. मॉडेल सेव्ह करणे (Model Saving)
# प्रशिक्षित मॉडेलला 'mushroom_model.joblib' या नावाने सेव्ह करणे
joblib.dump(model, 'mushroom_model.joblib')
print("\nमॉडेल 'mushroom_model.joblib' या नावाने यशस्वीरित्या सेव्ह झाले आहे.")
```

```python
from flask import Flask, request, jsonify
import joblib
import pandas as pd

# Flask ॲप सुरू करणे
app = Flask(__name__)

# सेव्ह केलेले मॉडेल आणि कॉलम्स लोड करणे
try:
    model = joblib.load('mushroom_model.joblib')
    model_columns = joblib.load('model_columns.joblib')
    print("मॉडेल आणि कॉलम्स यशस्वीरित्या लोड झाले.")
except Exception as e:
    print(f"मॉडेल लोड करताना त्रुटी आली: {e}")
    model = None
    model_columns = None

@app.route('/predict', methods=['POST'])
def predict():
    if model is None or model_columns is None:
        return jsonify({'error': 'मॉडेल लोड झालेले नाही, कृपया सर्वर लॉग तपासा.'}), 500

    # रिक्वेस्टमधून JSON डेटा मिळवणे
    json_data = request.get_json()
    if not json_data:
        return jsonify({'error': 'रिक्वेस्टमध्ये डेटा नाही.'}), 400

    # JSON डेटाला DataFrame मध्ये रूपांतरित करणे
    input_df = pd.DataFrame([json_data])

    # आलेल्या डेटाचे वन-हॉट एन्कोडिंग करणे
    input_encoded = pd.get_dummies(input_df)

    # मूळ मॉडेलनुसार कॉलम्स जुळवणे (Reindexing)
    # जे कॉलम्स इनपुटमध्ये नाहीत, ते 0 ने भरले जातील
    final_df = input_encoded.reindex(columns=model_columns, fill_value=0)

    # भविष्यवाणी करणे
    try:
        prediction = model.predict(final_df)
        probability = model.predict_proba(final_df)
    except Exception as e:
        return jsonify({'error': f'भविष्यवाणी करताना त्रुटी आली: {e}'}), 500


    # भविष्यवाणीच्या आधारावर परिणाम देणे (0 = Edible, 1 = Poisonous)
    if prediction[0] == 1:
        result = 'Poisonous (विषारी)'
    else:
        result = 'Edible (खाण्यायोग्य)'

    # JSON प्रतिसाद तयार करणे
    response = {
        'prediction': result,
        'probability': {
            'edible': round(probability[0][0], 4),
            'poisonous': round(probability[0][1], 4)
        }
    }
    return jsonify(response)

if __name__ == '__main__':
    app.run(debug=True, port=5000)
```
